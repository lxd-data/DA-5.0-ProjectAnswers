```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
customers = pd.read_csv("https://raw.githubusercontent.com/pathstream-curriculum/Python/master/Project-2/tco-customers.csv")
```

```python
churn = pd.read_csv("https://raw.githubusercontent.com/pathstream-curriculum/Python/master/Project-2/tco-churn.csv")
```

```python
customers.info()
customers.head(15)
```

```python
customer_subset = customers[["age", "gender", "number_of_referrals", "offer", "monthly_charge", "total_charges", "contract", "payment_method", "under_30", "senior_citizen", "married"]]
```

```python
gender_dict = dict(customer_subset.gender.value_counts())
x_labels = gender_dict.keys()
y_labels = gender_dict.values()
plt.bar(x_labels, y_labels)
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.title("Gender Breakdown in the Customer CSV")
```

```python
age_dict = dict(customer_subset.age.value_counts())
print(age_dict)
print(min(age_dict.values()))
print(min(age_dict.keys()))
print(max(age_dict.keys()))
print(max(age_dict.values()))
x_labels = age_dict.keys()
y_labels = age_dict.values()
plt.bar(x_labels, y_labels)
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.title("Age Breakdown in the Customer CSV")
```

```python
sns.lineplot(x="offer", y="monthly_charge", data=customer_subset)
```
