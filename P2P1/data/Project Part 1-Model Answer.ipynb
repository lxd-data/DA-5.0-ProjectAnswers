## <p style="color:red;">2A. Standardize Column Titles</p>

```python
# WRITE YOUR SOLUTION HERE. DO NOT DELETE. THIS CELL IS GRADED.
demographics.columns = demographics.columns.str.strip().str.replace(" ", "_").str.lower()
services.columns = services.columns.str.strip().str.replace(" ", "_").str.lower()
```

## 2B. Edit Data Types

```python
demographics.number_of_dependents.value_counts()
demographics = demographics.drop(list(demographics[demographics.number_of_dependents == 'O'].index))
demographics.number_of_dependents.value_counts()
```

## <p style="color:red;">2C. Locate & Fix Input Errors</p>

```python
demographics['married'].value_counts()
```

**The code cell below is graded. Do not delete the cell.**

```python
# WRITE YOUR SOLUTION HERE. DO NOT DELETE. THIS CELL IS GRADED.
demographics = demographics.replace("Y", "Yes")
demographics = demographics.replace("N", "No")
demographics.married.value_counts()
```

# 3. SERVICES DATASET WRANGLING

## 3A. Standardize Column Titles

```python
# already done above, heres same code
services.columns = services.columns.str.strip().str.replace(" ", "_").str.lower()
```

## 3B. Convert Data Types


```python
services.customer_enrollment_date = pd.to_datetime(services['customer_enrollment_date'])
```

# 4. MERGE THE DATAFRAMES

## 4A. Identify the Connecting Columns

```python
demographics['customer_id']
```

```python
services['customer_id']
```

## <p style="color:red;">4B. Manipulate the Connecting Columns</p>

**The code cell below is graded. Do not delete the cell.**

```python
# WRITE YOUR SOLUTION HERE. DO NOT DELETE. THIS CELL IS GRADED.
demographics['customer_id']= demographics['customer_id'].apply(lambda customer_id: customer_id.replace("TCO", "").replace("-", ""))
```

## <p style="color:red;">4C. Merge the DataFrames</p>
**The code cell below is graded. Do not delete the cell.**

```python
# WRITE YOUR SOLUTION HERE. DO NOT DELETE. THIS CELL IS GRADED.
final_df = demographics.merge(services, on = 'customer_id', how='outer') # including all instances
```
